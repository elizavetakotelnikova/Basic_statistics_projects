```python
import pandas as pd
import numpy as np
```

```python
booking = pd.read_csv("https://stepik.org/media/attachments/lesson/360344/bookings.csv", delimiter=";")
booking
```

```python
booking.dtypes
```

```python
#renaming columns
booking.columns = map(str.lower, booking.columns)
booking.columns = booking.columns.str.replace(" ", "_")
booking['succesful_booking'] = np.where(booking['is_canceled']
                       == 0, 1, 0)
booking
```

```python
#ranking top five countries with the most amount of succesful bookings
booking_by_country = booking.query("succesful_booking == 1").groupby('country').agg({'succesful_booking': 'sum'}) \
    .sort_values("succesful_booking", ascending=False)
booking_by_country.head()
```

```python
#another way to rank top five countries with the most amount of succesful bookings
booking.loc[booking.is_canceled == 0].country.value_counts().head(5)
```

```python
#finding a mean of stays_nights in different types of hotels
booking.groupby('hotel').stays_total_nights.mean().round(2)
```

```python
booking['total_kids'] = booking.children + booking.babies
booking
```

```python
booking.groupby('hotel').total_kids.mean().round(2)
```

```python
has_kids = (booking.query('total_kids != 0 and is_canceled == 1').shape[0] / booking.query('total_kids != 0').shape[0])
without_kids = (booking.query('total_kids == 0 and is_canceled == 1').shape[0] / booking.query('total_kids == 0').shape[0])
if (has_kids > without_kids):
    print('more canceled trips with kids')
else:
    print('more canceled trips without kids')
print(round(max(has_kids, without_kids) * 100, 2))
```
