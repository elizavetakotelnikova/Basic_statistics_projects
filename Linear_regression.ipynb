```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
URL = 'https://stepik.org/media/attachments/lesson/8086/states.csv'
df = pd.read_csv(URL)
df
```

```python
plt.scatter(df['hs_grad'], df['poverty'])
plt.xlabel("hs_grads (%)")
plt.ylabel("people living in poverty (%)")

# set the title of the chart name
plt.title("Scatter plot for poverty and education")
plt.show()
```

```python
x = df['hs_grad']
y = df['poverty']

gradient, intercept, r_value, p_value, std_err = stats.linregress(x, y)
plt.plot(x, intercept + gradient*x, color='red')
plt.plot(x, y,'ob') 
dt1 = df[['hs_grad', 'poverty']]
plt.show()
stats.linregress(x, y)
```

```python
# y = -0.62*hs_grad + 64.78
intercept = 64.78
slope = -0.62
```

```python
stats.f_oneway(x, y)
```

```python
#Residuals plot
import seaborn as sns
sns.residplot(x='hs_grad', y='poverty', data=df)
plt.show()
```

```python
#QQ plot to check homoscedasticity
from statsmodels.api import qqplot
residuals = y - (intercept + slope * x)
qqplot(data=residuals, fit=True, line='45')
plt.show()
```

We can draw a conclusion, that in general homoscedascity is okay for this data and linear regression can be implement. But there are few states that has a positive deviation from the predicted value

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
ax.scatter(xs=df['hs_grad'], ys=df['white'], zs=df['poverty'])
ax.set_xlabel('Higher education(%)')
ax.set_ylabel('White(%)')
ax.set_zlabel('Poverty(%)')

plt.show()
```

```python
from sklearn.linear_model import LinearRegression
import statsmodels.formula.api as smf

def predicted_value(x, y):
  return (lm.params.Intercept + lm.params.white * x + lm.params.hs_grad * y)
  
x = df[['white', 'hs_grad']]
y = df['poverty']
regression = LinearRegression().fit(x, y)
lm = smf.ols(formula='poverty ~ white + hs_grad', data=df).fit()

#plane of predicted values
d1, d2 = [], []
for x in np.linspace(min(df['white']), max(df['white']), 100):
    for y in np.linspace(min(df['hs_grad']), max(df['hs_grad']), 100):
        d1.append(x)
        d2.append(y)
d1 = np.array(d1).reshape(-1, 1)
d2 = np.array(d2).reshape(-1, 1)
plane = regression.predict(np.concatenate([d1, d2], axis=1))


fig = plt.figure(figsize=(10, 8))
ax = plt.axes(projection='3d')
ax.plot_trisurf(d2.ravel(), d1.ravel(), plane.ravel(), alpha=0.2)
ax.set_xlabel('Higher education(%)')
ax.set_ylabel('White(%)')
ax.set_zlabel('Poverty(%)')

data_above = df[df.poverty >= predicted_value(df.white, df.hs_grad)]
ax.scatter(data_above.hs_grad, data_above.white, data_above.poverty, c='g', marker='o')

data_below = df[df.poverty < predicted_value(df.white, df.hs_grad)]
ax.scatter(data_below.hs_grad, data_below.white, data_below.poverty, c='r', marker='o')

plt.show()
```

```python
multiple_regression = smf.ols(formula="poverty ~ metro_res + white + hs_grad + female_house", data=df).fit()
print(multiple_regression.summary())
```

### Using a multiple regression model, we can see how each factor is associated with a percentage of people living in poverty. As we can see, statistically significant correlation is between a poverty and location, higher education and the percentage of white population in the state. In each case a correlation is negative.
